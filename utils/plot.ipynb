```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
costs_file_path = f"{os.path.abspath("")}/../data/costs.csv" 
cost_g_file_path = f"{os.path.abspath("")}/../data/cost_g.csv" 
cost_r_file_path = f"{os.path.abspath("")}/../data/cost_r.csv" 
cost_c_file_path = f"{os.path.abspath("")}/../data/cost_c.csv" 
thetadot_file_path = f"{os.path.abspath("")}/../data/thetadot.csv" 
theta_file_path = f"{os.path.abspath("")}/../data/theta.csv" 
```

```python
costs = np.genfromtxt(costs_file_path, delimiter=',')
cost_g = np.genfromtxt(cost_g_file_path, delimiter=',')
cost_r = np.genfromtxt(cost_r_file_path, delimiter=',')
cost_c = np.genfromtxt(cost_c_file_path, delimiter=',') 

thetadot = np.genfromtxt(thetadot_file_path, delimiter=',')
theta = np.genfromtxt(theta_file_path, delimiter=',')
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4))
fig.suptitle('Costs', fontsize=16)
axs = axs.flatten() 

axs[0].plot(costs)
axs[0].set_title("Total cost over CEM iterations")
axs[0].set_xlabel("Iteration")
axs[0].set_ylabel("Cost")

axs[1].plot(cost_g)
axs[1].set_title("Goal cost form last CEM iteration")
axs[1].set_xlabel("Step")
axs[1].set_ylabel("Cost")

axs[2].plot(cost_r)
axs[2].set_title("Orientation cost form last CEM iteration")
axs[2].set_xlabel("Step")
axs[2].set_ylabel("Cost")

plt.tight_layout()
# plt.savefig("polynomials.png")
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Trajectory', fontsize=16)
axs = axs.flatten() 

axs[0].plot(thetadot)
axs[0].set_title("Joint Velocities")
axs[0].set_xlabel("Step")
axs[0].set_ylabel("Velocity")
axs[0].legend(['joint 1', 'joint 2', 'joint 3', 'joint 4', 'joint 5', 'joint 6'], loc='upper left')


axs[1].plot(theta)
axs[1].set_title("Joint Angles")
axs[1].set_xlabel("Angle")
axs[1].set_ylabel("Step")
axs[1].legend(['joint 1', 'joint 2', 'joint 3', 'joint 4', 'joint 5', 'joint 6'], loc='upper left')


plt.tight_layout()
# plt.savefig("polynomials.png")
plt.show()
```
