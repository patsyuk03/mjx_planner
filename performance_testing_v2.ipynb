```python
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
```

```python
SOURCE_DIR = os.path.abspath("")

data = {
    'batch_size': [500, 1000, 1500, 2000, 2500, 3000, 3500, 4000, 4500, 5000],
    'num_steps': [8, 13, 18, 23, 28, 33, 38, 43, 48, 53],
    'num_iter': [1, 6, 11, 16, 21, 26, 31, 36, 41, 46]
}

dirs = ['batch_size', 'num_steps', 'num_iter']

compilation_dir = list()
computation_dir = {
    'batch_size': [],
    'num_steps': [],
    'num_iter': []
}
for _dir in dirs:
    compilation_dir.append(f"{SOURCE_DIR}/data/performance_2/compilation_{_dir}.csv") 
    computation_folder_dir = f"{SOURCE_DIR}/data/performance_2/{_dir}" 

    for i in data[_dir]:
        computation_dir[_dir].append(f'{computation_folder_dir}/computation_{i}.csv')

    print(compilation_dir[-1])
    for i in computation_dir[_dir]:
        print(i)
```

```python
dirs = ['batch_size', 'num_steps', 'num_iter']

dt_compilation = list()
for _dir in compilation_dir:
    dt_compilation.append(np.genfromtxt(_dir, delimiter=','))

dt_computation = {
    'batch_size': [],
    'num_steps': [],
    'num_iter': []
}

for _dir in dirs:
    for i in computation_dir[_dir]:
        dt_computation[_dir].append(np.genfromtxt(i, delimiter=','))

    computation_dir[_dir] = np.array(computation_dir[_dir])



print(dt_compilation)
print(dt_computation)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 3))
fig.suptitle('Peftormance', fontsize=16)
axs = axs.flatten() 

labels = ['Batch size', 'Number of steps', 'Number of iterations']
fsize = 16


i=0
mean = np.mean(dt_computation['batch_size'], axis=-1)
std = np.std(dt_computation['batch_size'], axis=1)
print(dt_computation['batch_size'])
axs[i].plot(data['batch_size'], mean, 'g')
axs[i].fill_between(data['batch_size'], mean - std, mean + std, color='g', alpha=0.3)
axs[i].set_xlabel("Batch size", fontsize=fsize)
axs[i].set_ylabel("Compute Time (s)", fontsize=fsize)
axs[i].tick_params(axis='both', labelsize=12)
axs[i].grid(True)

i+=1
mean = np.mean(dt_computation['num_steps'], axis=-1)
std = np.std(dt_computation['num_steps'], axis=1)
axs[i].plot(data['num_steps'], mean, 'b')
axs[i].fill_between(data['num_steps'], mean - std, mean + std, color='b', alpha=0.3)
axs[i].set_xlabel('Number of steps', fontsize=fsize)
axs[i].tick_params(axis='both', labelsize=12)
axs[i].grid(True)

i+=1
mean = np.mean(dt_computation['num_iter'], axis=-1)
std = np.std(dt_computation['num_iter'], axis=1)
axs[i].plot(data['num_iter'], mean, 'r')
axs[i].fill_between(data['num_iter'], mean - std, mean + std, color='r', alpha=0.3)
axs[i].set_xlabel('Number of iterations', fontsize=fsize)
axs[i].tick_params(axis='both', labelsize=12)
axs[i].grid(True)


plt.tight_layout()
# plt.savefig(f"{os.path.abspath("")}/img/performance_2/perf_2.png" )
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
fig.suptitle('Peftormance', fontsize=16)
axs = axs.flatten() 

labels = ['Batch size', 'Number of steps', 'Number of iterations']


i=0
for key, _dt_computation in dt_computation.items():
    axs[i].plot(np.mean(_dt_computation, axis=-1))
axs[i].set_title("CEM Compute Time")
axs[i].set_xlabel("X")
axs[i].set_ylabel("Compute Time (s)")
axs[i].legend(labels, loc='upper left')

i+=1
for _dt_compilation in dt_compilation:
    axs[i].plot(_dt_compilation)
axs[i].set_title("CEM Compilation Time")
axs[i].set_xlabel("X")
axs[i].set_ylabel("Compilation Time (s)")
axs[i].legend(labels, loc='upper left')


plt.tight_layout()
plt.savefig(f"{os.path.abspath("")}/img/performance_2/perf.png" )
plt.show()
```

```python
values = {
    "X": np.arange(0, 10),
    "Batch size": data['batch_size'],
    "Number of steps": data['num_steps'],
    "Number of iterations": data['num_iter'],
}

df = pd.DataFrame(values)
df_transposed = df.T

fig, ax = plt.subplots(figsize=(10, 1.5)) 
ax.axis('off') 
ax.set_position([0.2, 0, 0.78, 1])
table = ax.table(cellText=df_transposed.values,
                 rowLabels=df_transposed.index,
                 loc='center',
                 cellLoc='center',
                 rowLoc='center')

for col in range(df_transposed.shape[1]):
    cell = table[(0, col)]
    cell.set_text_props(weight='bold')

for key, cell in table.get_celld().items():
    if cell.get_text().get_text() == "X":
        cell.set_text_props(weight='bold')


table.scale(1, 1.5)
plt.savefig(f"{os.path.abspath("")}/img/performance_2/time_perf_table.png")
plt.show()
```
